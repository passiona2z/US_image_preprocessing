### sample 만들기

```python
import glob
import random
```

```python
li = glob.glob('/dshome/ddualab/seonjin2/denosing/dataset/fatty liver')

img_list = []

for i in li :
    
    img = glob.glob(i + '/*.png')
    img_list.extend(img)
```

```python
print(len(img_list))
random.seed(42)
samplelist = random.sample(img_list, 5)
print(len(samplelist))
samplelist
```

```python
import shutil

for i in samplelist :
    shutil.copy(i, '/dshome/ddualab/seonjin2/denosing/report/sample/fatty liver')
```

### mat 파일

```python
mat = scipy.io.loadmat('/dshome/ddualab/seonjin2/denosing/dataset/fatty liver_raw/dataset_fatty liver.mat')
```

```python
mat['data'].shape
```

```python
arr = mat['data'][0][0][3]
```

```python
arr.sum()
```

```python
import numpy as np

for i in range(1,55) :

    data = mat['data'][0][i][3]
    arr = np.concatenate((arr, data), axis=0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.imshow(arr[0], cmap='gray')
plt.show()
```
